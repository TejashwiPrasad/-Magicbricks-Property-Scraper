```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url='https://www.magicbricks.com/flats-in-mumbai-for-sale-pppfs'
```

```python
r=requests.get(url)
r
```

```python
htmlcontent=r.content
htmlcontent
```

```python
soup=bs(htmlcontent,'html.parser')
soup
```

```python
print(soup.prettify())
```

# TITLE

```python
title=soup.title
print(title.text)
```

# FLAT NAMES

```python
f_n=soup.find_all('h2',class_='mb-srp__card--title')
```

```python
len(f_n)
```

```python
flat_name=[]
for i in range(0,len(f_n)):
    flat_name.append(f_n[i].get_text())
len(flat_name)
```

```python
flat_name
```

# FLAT RATE

```python
rate=soup.find_all('div',class_='mb-srp__card__price--amount')
```

```python
flat_rate=[]
for i in range(0,len(rate)):
    flat_rate.append(rate[i].get_text())
len(flat_rate)
```

```python
flat_rate
```

# OWNER/BUILDER NAME

```python
build=soup.find_all('div',class_='mb-srp__card__ads--name')
```

```python
builder=[]
for i in range(0,len(build)):
    builder.append(build[i].get_text())
len(builder)
```

```python
builder
```

# FLAT AREA IN SQUARE FEET

```python
P_sq=soup.find_all('div',class_='mb-srp__card__price--size')
len(P_sq)
```

```python
Per_sq=[]
for i in range(0,len(P_sq)):
    Per_sq.append(P_sq[i].get_text())
len(Per_sq)
```

```python
Per_sq
```
